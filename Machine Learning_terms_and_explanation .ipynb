#### REGRESSION TECHNIQUE ####

1. **Mean** => Overall average value of N inputs

2. **Standard Deviation** => Square Root of 1/N Summation(i=0 to N) (x(i) - mean)^2 it is also called **Population SD**
                          **Sample SD** => Square Root of 1/N summation(i=0 to N-1) (x(i)-mean)^2
                          
3. **covarience** => Is a measure of  much **two random variables** vary together.
          Cov(x,y) = Summation(i=0 to N) (x(i)-mean(x)) * (y(i)-mean(y)) / N-1
          
4. **corelation** => Corelation between Two data

5. **Simple Linear Regression** => Linear Relationship to modeling the scalar response one or more responsible variables
                        y = a + b * x
                    x => input_data, {a,b} => someconstantvalue, y=>output
                    
6. **Ordinary Least Square** =>  procedure seeks to minimize the sum of square resuidals. This means given a regression line we calculate **distance from datapoint to the regression line**, square it, and sum of all square error together. For **Multiple Linear Regression** find the **Regression Plane** and calculate sum of squares error calculation.

7. **Mean squared Error(MSE)** => Way to measure how well its predictions actually match the observed data.
                            = 1 / N (Summation(y(i)-f(x(i))^2) where range is 1 to N
                        
                        
8. **Coordinate Descent** => Derivative that minimization of a multivariable function F(x) along one direction at a time. To optimize the network, Choose one coordinate(randomly) and iterate through all the N data.



9. **sum of squared Error(SSE)** => 1/2 sum(actual_data - predicted_data)^2. ie.., here predicted_data is nothing but a linear regression line.



10. **Gradient Descent** => [https://www.kdnuggets.com/2017/04/simple-understand-gradient-descent-algorithm.html] Refer this link for better understanding

### OVER FITTING AND UNDER FITTING ###

1. **OverFitting** => Model trained too well, and gives best accuracy. Best fails to predict the class when it comes to new data
2. **UnderFitting** => Vice versa to overfitting. The model performs poor both on Training data and new data. 



3. **Cross Validation** => splitting our data into K subsets, and train on one of **k-1 subsets** and validate the model on rest k model. There are two types of cross validation
           3.1 **KFold validation** => split our data into K subset and train on K-1 subset and leave the last subset. We then average each of the hold and validate our model
           3.2 **Leave One Out Cross Validation**

### CLUSTERING ###

Following are types of Clusters

1. Hierarchical Clustering  => also known as **nesting clustering** as it also clusters to exists within bigger clusters to form a tree

2. Partition Clustering => Division of set of data objects into non-overlapping clusters such that each object is exactly one subset

3. Exclusive Clustering => They assign each value to single cluster

4. Overlapping Clustering => Used to reflect the fact than an object can simultaneously belong to more than one group

5. Fuzzy Clustering => Every object belong to every cluster with a membership weight that goes between 0: if its absolutely belong to the cluster, 1: if it completly belongs to the cluster

6. Complete Clustering => It perform **Hierarchical Clustering** using a set of dissimilarities on 'n' objects that are being clustered.
